# 2) Data Cleaning

### 2a) Check for any Null Values and remove.

Since we know there are 5 rows. Following code, we will identify the which rows consists Null value and remove specifically.

```python
music[music.isnull().any(axis=1)]
```

After identify its row 10000-10004, lets remove them accordingly.

```python
music.drop([10000, 10001, 10002, 10003, 10004], inplace = True)
```

Next, we would like to check for any duplication rows in our dataset.

```python
music[music.duplicated()]
```

There isn't, we will move on.

In the dataset, we have also spotted some rows under column "Duration_ms" containing values of -1. This is invalid value, a song duration would never be -1, which may produce errors in our findings, hence we remove them

```python
music.drop(music[music['duration_ms'] == -1.0].index, inplace = True)
music
```

Besides duration_ms, "tempo" turns out to have invalid value as well "?", These values need to either be removed or replaced with median or mean value. If not errors may occur during our findings. 

```python
music[music["tempo"] == "?"]
```

There exists 4501 rows with tempo value "?", this problem is significant. Hence, need to be removed.

```python
music.drop(music[music['tempo'] == "?"].index, inplace = True)
music
```

The reason why we are removing invalid values instead of replacing it with median or mean values is because simplifies the data and makes it easier to work with and bias may happen if the mean or median value is inappropriate or if the distribution of the data is skewed.

```python
#Doing a final check on the dataset
pd.isnull(music).sum() #Check for any remaining Null Values 
```

# Remove unwanted columns

In the dataset, there exist some unwanted column such as "obtained_date", "instance_id", "track_name" and "artist_name". These columns do not have any relation with the song genre, a date, id and track_name doesn't determine a song's genre and wouldn't be able to help us with the prediction for song genre. And same artist may produce a different types of genre, we cannot interpret anything based on the artist name. Hence we are removing these columns."

```python
music = music.drop(["instance_id", "track_name", "obtained_date", "artist_name"], axis = 1)
music.head()
```

The dataset index was reset since 5 rows were removed earlier 

```python
music.reset_index(inplace = True)
```
